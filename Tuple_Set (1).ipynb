## Tuple in Python


Defination:
     Tuple is a built-in data type in Python used to store multiple items in a single variable.
A tuple is written using round brackets ( ), and the elements are separated by commas.

```python
#Create a tuple
a=10,20,30
b=()
d=(1,) # Single element tuple (comma is important)

print(type(a))
print(type(d))
print(type(b))
```

### Properties of Tuple



```python
Ordered
Heterogeneous
Nestead
Duplicates allow
Immutable
```

```python
t=23,46,"liki",50,["liki",2]   # Tuple is orderd 
t
```

```python
k=23,"hii",[23,"python"],2.4,True
k                                    #Heterogeneous allow all datatypes 
```

```python
k=("jii",("python","hii",23))   # Nestead tuple accept 
k
```

```python
l=(23,4,4,5,23,"python","python")
l                       # Duplicates allow 
```

```python
t=(12,23,4,"hii","python")
t[0]                 #After creating a tuple, you cannot change its values.
t[1]="Hello"   #tuples are immutable, they use less memory and are faster than lists.
```

```python
#built in tuple

t=(1,2,3,4,1)
print(t.count(1))               # Only 2 built in methods in tuple 
print(t.index(2))
```

```python
#Built in Functions in tuple
t=(1,2,3,4,1)
print(len(t))
print(min(t))
print(max(t))
print(sum(t))
del t
print(t) #Tuple elements cannot be modified or deleted individually, but the whole tuple variable can be deleted using del.
```

## Set in Python

Defination:
  A set is a built-in data type in Python used to store multiple items in a        single variable.
  Sets are written using curly brackets { }

```python
#Create a set
a={12,13,14} 
print(type(a))
b={12}
print(type(b))
c={}  #empty curly brackets means it takes dictionary not a set one value is requered 
print(type(c))
```

## Properties of set

Not Ordered
Heterogeneous
Not allow Nestead
Duplicates not allow
mutable

```python
s={20,30,"python","hii"}   
s               # Set is unordered
```

```python
d={20,30,"python","hii",["liki",45],(23,45,"hello")}
d   #Accept hashable(immutable) datatypes only(like tuple,string,int..)
```

```python
a={1,2,3,{5,6}}
a                 # Not allow Nestead
                 #All elements inside a set must be hashable (immutable).
```

```python
a={23,33,23,444,55,23}
a                               #Duplicates not allow, set automatically remove duplicates 
```

```python
a={23,33,23,444,55,23}
print("Before:",id(a))    #Same ID,Object modified in same memory
a.add("Python")           # So Set is mutable
print("After:",id(a))
```

```python
#SET
a={1,2,3,90,'kl'}
a.add((12,13,"hii")) 
a
```

```python
a={1,2,3,90,'kl'}
a.update(["liki",23,2002])
a
```

```python
Observation:
     1.add() inserts only one single element into the set.
     2.update() inserts multiple elements from an iterable (list, tuple, set, string).
```

```python
Question:
there are 2 classrooms in a school classA consistes of karan,shirisha,krish,
nagesh,saurabh and classB consist of sakshi,shrikant,sruthi,srishti . 
There was a competition in the school and students scored following 
marks in the competition. 
In competition1 karan,sruthi,sakshi scored 3.4,3.5,4.7 
and in competition2 karan,shirisha,nagesh,saurabh scored 12,13,14,15

how many students in classA only participated in competition2
how many students participated in only 1 competition
how many students participated in both the competition who were 
not part of classA
total number of students in school
total number of students in the competition
toal number of students in the competition who were not from class A
```

```python
classA = {"karan", "shirisha", "krish", "nagesh", "saurabh"}
classB = {"sakshi", "shrikant", "sruthi", "srishti"}

comp1 = {"karan", "sruthi", "sakshi"}
comp2 = {"karan", "shirisha", "nagesh", "saurabh"}

```

```python
#1.How many students in Class A only participated in Competition 2?
((classA & comp2) - comp1)
```

```python
#2 How many students participated in only 1 competition?
comp1 ^ comp2
```

```python
#3 How many students participated in both competitions who were NOT part of Class A?
(comp1 & comp2) - classA
```

```python
#4 Total number of students in school
classA | classB
```

```python
#5 Total number of students in competition
comp1 | comp2
```

```python
#6 Total number of students in competition who were NOT from Class A
(comp1 | comp2) - classA
```
