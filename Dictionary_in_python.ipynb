## Methods and functions on Dictionary

```python
1. get()
2. keys()
3. values()
4. items()
5. update()
6. pop()
7.popitem()
8.clear()
9. setdefault()
10. copy()
11. fromkeys()
12.len(d)
13.sorted(d)
14.max(d)
15.min(d)
```

```python
#get()
d = {"name": "Likitha", "age": 22}
print(d.get("name"))         #get returns the value of the particular key
print(d.get(d.get("city")))  # if the value is not found get returns the None 
print(d.get("city","NA"))   #provided default value "NA"
print(d)
```

```python
#keys()
d = {"name": "Likitha", "age": 22,"course":"python"}
d.keys()               #provides all keys in given dictionary
```

```python
#values()
d = {"name": "Likitha", "age": 22,"course":"python"}
d.values()   #provides all values in given dictionary
```

```python
#items()
d = {"name": "Likitha", "age": 22,"course":"python"}
d.items()   #Returns key-value pairs as tuples.
```

```python
#update()
l = {"name": "Likitha", "age": 22,"course":"python"}
l.update({"city": "Hyderabad"})  #If key does NOT exist → update() adds it.
l.update({"age": 25}) 
l                     #If key already exists → update() changes the value.
```

```python
#pop()
l={"apples":3,"cherry":2,"mango":5}
l.pop() #pop expected at least 1 argument, got 0 , in dictionary it is not delete the last key and values
```

```python
l={"apples":3,"cherry":2,"mango":5}
print(l.pop("mango"))
print(l.pop("hii")) #if key is not found it gives Error
```

```python
#popitem()
d = {"name": "Likitha", "age": 22, "city": "Hyderabad"}
item = d.popitem()     #popitem gives the last inserted key-value pair
print(item)              #Returns it as a tuple (key, value)
```

```python
#clear()
d = {"name": "Likitha", "age": 22, "city": "Hyderabad"}
d.clear()   #Removes all key-value pairs from dictionary
print(d)      #Does NOT delete dictionary variable
```

```python
#setdefault()
d = {"name": "Likitha"}    #"city" was not there Added automatically becuase of setdefault()
result = d.setdefault("city", "Hyderabad")
d
```

```python
d = {"name": "Likitha", "age": 22}
result = d.setdefault("name", "Unknown")
print(result)  #If key exists → returns its value
print(d)
```

```python
#copy()
d = {"name": "Likitha", "age": 22}
d2 = d.copy()  #Makes a new dictionary with same key-value pairs
print(d2)
```

```python
d = {"a": 10, "b": 20}
d2 = d.copy()
d2["a"] = 100
print(d)
print(d2)   #Changing copy does NOT affect original
```

```python
d={"liki": [1, 2, 3]}
d2=d.copy()
d2["liki"].append(4)
print(d)
print(d2)
```

```python
keys = ["a", "b", "c"]
d = dict.fromkeys(keys, 2)
print(d)
```

```python
keys = ["a", "b"]
d = dict.fromkeys(keys, []) #If value is mutable (like list), all keys share same object.
d["a"].append(1)
print(d)
```

```python
keys = ["x", "y", "z"]
d = dict.fromkeys(keys)  #Not provide any value means Default value is None 
print(d)
```

```python
#len

d = {"name": "Likitha", "age": 22, "city": "Hyderabad"}
print(len(d)) #Returns the number of key-value pairs in dictionary.
```

```python
#sorted(d)
d = {"b": 2, "a": 1, "c": 3}
print(sorted(d)) #Sorts the keys of the dictionary
```

```python
#max()
d = {"a": 10, "c": 30, "b": 20}
print(max(d)) #Returns the largest key in the dictionary
```

```python
d = {5: "A", 2: "B", 9: "C"}
print(max(d))   #Largest number in key
```

```python
#min()
d = {"a": 10, "c": 30, "b": 20}
print(min(d))   #Returns the smallest key in the dictionary
```

```python
d = {5: "A", 2: "B", 9: "C"}
print(min(d))
```
