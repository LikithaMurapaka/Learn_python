Observations:
1.A list is ordered.It supports indexing and slicing, and index starts from 0.
2.A list allows all data types, including immutable data types.It can store different types of elements together, so it is called heterogeneous.
3.A list is mutable, meaning we can create and manipulate (add, remove, update) elements after creation.
If the id remains the same after modification, it means the object is mutable.

```python
l=[10,20,30,"liki"]
l.append((12,13,"apple"))   #append() adds one element to the end of the list.
l                           #It adds the item as a single element.
```

```python
l=[10,20,30, "liki"]
l.extend((12,13,"apple"))   #extend() adds multiple elements to the list.
l

# append() adds the whole object as one element, whereas extend() adds each element separately.
```

```python
a=[90,80,70,0]   #removes values from the end
a.pop()    
print(a)

a.pop(0)
print(a)       #can delete values at exact location


x=a.pop(0)
print(x)       #It returns the removed value.
```

```python
a=[90,80,70,0] 
a.remove(70)      #accepts value and deletes them
a

a = [10, 20, 30, 20]
a.remove(20)
print(a)                  #Deletes the first occurrence of that value.

                          
a = [10, 20, 30, 20]
x=a.remove(20)
print(x)             #Does not return anything.
```

```python
a.clear() 
a                      #removing all elements in a list 
```

```python
m=["liki",24,45]
del(m) #removes the entire value with the variable
m

#del(m) deletes the entire variable m, not just the values.

#The variable name m is completely removed from memory.        
```

```python
a = [10, 20, 30, 20]    #Returns the index position of the first occurrence of a value.
print(a.index(20))     #Gives ValueError if value not found.
```

```python
a = [10, 20, 30, 20]     #Returns the number of times a value appears in the list.
print(a.count(20))
```

```python
a = [4, 2, 1, 3]
a.sort()              #Sorts the list in ascending order by default.
print(a)             #Modifies the original list.
```

```python
a = [1, 2, 3]
a.reverse()    #Reverses the order of elements in the list.
print(a)       #Does not sort, just reverses current order.
```

```python
a = [[1, 2], 3]
b = a.copy()        # copy Creates a new outer list.
print(b)

a[0][0] = 100
print(b)
print(id(a) == id(b)) #It does not create new copies of nested objects.
```

```python
import copy
a = [[1, 2], [3, 4]]
b = copy.deepcopy(a)
a[0][0] = 100

print("Original:", a)
print("Copied:", b)
print(id(a) == id(b))   
```

```python
Observation: 
Copy :
creates a new outer list
Inner (nested) objects are shared
Changes inside nested objects affect both lists  

Deep Copy:
Deep copy Creates a new outer list
Creates new inner objects
Completely independent copy
```

```python
l=[20,40,56,"likitha","python"]
print(len(l))
```

```python
l = [23, 56, 70, 200, 34]  
print(min(l))   # Returns the smallest element in the list
print(max(l))   # Returns the largest element in the list
print(sum(l))   # Returns the sum of all elements in the list
```
