Rules for Naming Variables in Python

1. A variable name must start with a letter (a–z, A–Z) or an underscore        (_).
2. A variable name cannot start with a number.
3. Variable names cannot contain spaces.
4. Special characters are not allowed except underscore (_).
5. Variable names are case sensitive (age and Age are different).
6. There is no strict limit on the length of a variable name, 
   but it should be meaningful and readable.
7. Variable names should not use Python keywords

### Python Type Conversion

Type Conversion means changing one data type into another data type.In python there are 2 types of type conversions.
1.Implicit Type Conversion (Python does automatically)
2.Explicit Type Conversion (We do manually using functions like int(), float(), str())

```python
a = 10      # int
b = 5.5     # float

c = a + b   # int automatically converts to float
c
```

```python
a = 10
b = float(a)
print(b)
```

```python
a = 100
b = str(a)
print(b)
print(type(b))
```
