## Strings

A string is a data type used to represent textual data. It is an ordered sequence of characters, where each character can be a letter, digit, symbol, or whitespace.

Strings are immutable, meaning once a string is created, its content cannot be changed. Any operation that appears to modify a string actually creates a new string in memory.

Strings preserve order, allowing characters to be accessed in a defined sequence. This makes them suitable for tasks such as text processing, data cleaning, and pattern matching.

They support a wide range of built-in operations and methods for manipulation, such as searching, slicing, formatting, and transformation, without altering the original string.

Strings are widely used to store:

Names and labels

Messages and descriptions

File paths and URLs

Textual data from files, databases, and APIs

strings are essential for handling and manipulating human-readable data in programs.

### Observation

1)Strings are ordered.

2)Strings are homogeneous: they take all data types but are processed   character-wise, so they are not heterogeneous.

3)Strings are immutable: any operation that modifies a string creates a new string object, leaving the original string unchanged.

4)The memory address (id) of the new string is different from the original string.Changes are temporary in the new string; the original string is not affected.

5)Strings are iterable and support indexing and slicing.
