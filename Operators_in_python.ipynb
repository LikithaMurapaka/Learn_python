# Python Operators and Types 

Defination:An operator is a special symbol or keyword in Python that performs a specific operation on one or more operands (values or variables) and produces a result.

There are mainly 7 types:

1.Arithmetic Operators

2.Assignment Operators

3.Comparison (Relational) Operators

4.Logical Operators

5.Bitwise Operators

6.Membership Operators

7.Identity Operators 

## 1.Arithmetic Operators(+,-,*,**,/,//,%)

```python
a=10
b=20
print(a+b)   #add two numbers
print(a-b)   # subtraction 
print(a*b)   #multiplication
```

```python
#important differnce between (/,//,%)

a=20
b=3
print(a/b)      # Division operator provides the "quotient" and it all ways provides the float(decimal) value.
print(a//b)     # Floor Division **It removes the decimal part and rounds down to the nearest integer.(floor value)**
print(a%b)      # Modulo division provides the remainder
```

##  2.Assignment Operators (=,+=,-=,*=,/=)

```python
a=20
b=50

b=a                      # Assign value of a to b
print(b)                  # Output: 20

a+=10                    # Add 10 to a → a = a + 10   
print(a)                   # Output: 30  


b-=30                  # Subtract 30 from b → b = b - 30
print(b)                 # Output: -10

a*=10                    # Multiply a by 10 → a = a * 10
print(a)                   # Output: 300

b/=4                      # Divide b by 4 → b = b / 4
print(b)                   # Output: -2.5 
```

## 3.Relational (Comparison) Operators (>,<,<=,>=,==,!=)

```python
# All Relational operators return Boolean(True,False)
n1=30
n2=40
print(n1 < n2)   # True  → 30 is less than 40
print(n1 > n2)   # False → 30 is not greater than 40
print(n1 <= n2)  # True  → 30 is less than or equal to 40
print(n1 >= n2)  # False → 30 is not greater than or equal to 40
print(n1 == n2)  # False → 30 is not equal to 40
print(n1 != n2)  # True  → 30 is not equal to 40
```

## 4.Logical Operators(AND,OR,NOT)

```python
a=True
b=False

print(a and b)   # False → because both are not True
print(a or b)    # True  → because at least one is True
print(not a)     # False → opposite of True
```

## 5.Bitwise Operators (&,|,^,~,<<,>> )


```python
print(bin(15))    # It provides the binary represantion 
```

```python
print(7&10)    #Compares each bit of two numbers.
            #Returns 1 if both bits are 1, else 0
```

```python
print(12|8)   #if at least one bit is 1, else 0.
```

```python
print(~10)  #-(n+1)---->to perform negation using bitwise operator
```

```python
print(12>>3)    #Shifts bits to the right by n positions.
                #Removes bits from the right.
                #floor divides the number by 2^n
```

```python
print(12<<3)   #Shifts bits to the left by n positions
            #Adds zeros on the right.: multiplies the number by 2^n
```

```python
print(10^7)  #dissimilar bits will give me 1 as an answer
             #if bits are different(1) else 0
```

## 6.Membership Operators(in, not in)

```python
#Membership operators work with strings, lists, tuples, sets, dictionaries (keys), etc.
s = "Krishna Vamsi"
print("V" in s)      # True → "V" exists in the string
print("V" not in s)  # False → "V" is present, so "not in" is False
```

```python
n=123
print(3 in n)
#Membership operators are perfect for checking sequences.(iterable datatypes only)
```

## 7.Identity Operators (is ,is not)

```python
# is checks whether two variables point to the same object in memory.It checks object identity, not value equality.
```

```python
a=10
print(id(a))     #Python caches small integers (typically -5 to 256) for efficiency. So 10 points to the same object.
b=10              #same memory location so True            
print(id(b))    
a is b
```

```python
a='manisha is' 
print(id(a))            #coming to strings 
b='manisha is'          #Python interns strings that look like identifiers (no spaces, letters only) for memory optimization.
print(id(b))            #Strings with spaces, special characters, or dynamically created may not be interned automatically → stored in different memory locations.
a is b
```

```python
a=1000
print(id(a))
b=1000               #Larger integers(out of 256 so different memory addresses)
print(id(b))
a is b                
```

```python
a=10
b=10
a is not b
```
